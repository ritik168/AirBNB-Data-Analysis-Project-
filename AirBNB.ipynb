# Handling Missing Values 

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review':df['last review'].min()},inplace=True)
```

```python
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=["license","house_rules"],errors='ignore')
```

```python
df.head()
```

```python
# Remove dollar sign and convert to float
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

# Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

# Data Visulaization

# Q.1 What is the distribution of listing price?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title('Distribution of listing price')
plt.xlabel('Price $ ')
plt.ylabel('Frequency')
plt.show()
```

#### The histogram shows the fairly even distribution of listing price acroos different price renges,indicating no particular concetartion of listing in any price range. The KDE lines help visulaize this even spread more clearly, confirming that the datasets contains listing with a wide variety of price.

```python
df['room type']
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='room type',data=df,color='red')
plt.title('room type distribution ')
plt.xlabel('room type')
plt.ylabel('count')
```

# Q.3 How are listings distributed across different neighborhoods?

```python
plt.figure(figsize=(10,5))
sns.countplot(y='neighbourhood group',data=df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of Listing of Neighbourhood Group')
plt.xlabel('count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# Q.4 What is the relationship between price and room type?

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='room type',y='price',hue='room type',data=df)
plt.title('Price .vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price $')
plt.legend(title='Room Type')
plt.show()
```

# Q.5 How has the number of reviews change over with time?

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(10,5))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
